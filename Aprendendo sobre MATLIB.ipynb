```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [1,4,2,3]

figura = plt.figure(figsize=(16, 4))
figura.suptitle('dashboard em MATLIB')

figura.add_subplot(131)
plt.plot(x,y,label = 'dados')
plt.legend()
plt.title('grafico em linha')

figura.add_subplot(133)
plt.scatter(x,y)
plt.title('grafico em bolhas')

figura.add_subplot(132)
plt.bar(x,y)
plt.title('grafico em barras')


plt.show(figura)
```
